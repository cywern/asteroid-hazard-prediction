```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')
import seaborn as sns
```

```python
df=pd.read_csv('nasa.csv')
df.head() #Returns a new DataFrame containing the first n rows of the original DataFrame
```

```python
df.info() #Number of rows and columns in the DataFrame.
```

```python
df.shape #see the overall size of your data in terms of rows (observations) and columns (features)
```

```python
df = df.drop(['Neo Reference ID', 'Name', 'Orbit ID', 'Close Approach Date','Epoch Date Close Approach', 'Orbit Determination Date'] , axis = 1) #remove rows or columns from a DataFrame
df.head()
```

# ONE-HOT ENCODING OF HAZARDOUS COLUMN

### One-hot encoding is a technique used in machine learning to represent categorical variables as numerical features

```python
hazardous_labels = pd.get_dummies(df['Hazardous']) #This line essentially performs one-hot encoding on the "Hazardous" column
hazardous_labels
```

```python
df=pd.concat([df, hazardous_labels], axis=1) #This argument specifies that the concatenation should happen along axis 1, which means joining the DataFrames by columns
df.head()
```

```python
df = df.drop(['Hazardous'], axis = 1)
df.head()
```

```python
df.info()
```

# Observing Unique Values in Orbiting Body and Equinox

```python
df['Orbiting Body'].value_counts() #calculates the frequency or count of each unique value within the column
```

```python
df['Equinox'].value_counts()
```

```python
df = df.drop(['Orbiting Body', 'Equinox'], axis = 1) #dropped because both have only single unique value
```

# HEATMAP OF CORRELATION

### correlation = existence of a relationship between two variables

```python
plt.figure(figsize = (20,20))
sns.heatmap(df.corr(),annot = True)
```

```python
# Many Columns are there which are having correlation 1 because they are identical columns with a different unit. So, can be dropped
df = df.drop(['Est Dia in KM(max)', 'Est Dia in M(min)', 'Est Dia in M(max)', 'Est Dia in Miles(min)' ,'Est Dia in Miles(max)', 'Est Dia in Feet(min)', 'Est Dia in Feet(max)', 'Relative Velocity km per hr', 'Miles per hour', 'Miss Dist.(lunar)', 'Miss Dist.(kilometers)', 'Miss Dist.(miles)'], axis = 1)
df.head()
```

# FINAL HEATMAP

```python
plt.figure(figsize = (20,20))
sns.heatmap(df.corr(),annot = True)
```

# MODEL BUILDING

```python
x = df.drop([True], axis = 1)
y = df[True].astype(int)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y,random_state=0,test_size=0.3)
```

# XGBOOST CLASSIFIER

### XGBoost is a decision-tree-based ensemble Machine Learning algorithm that uses a gradient boosting framework. In prediction problems involving unstructured data (images, text, etc.) ... A wide range of applications: Can be used to solve regression, classification, ranking, and user-defined prediction problems

```python
pip install xgboost
```

```python
from xgboost import XGBClassifier
from matplotlib import pyplot
from xgboost import plot_importance

xbg_model = XGBClassifier()
xbg_model.fit(x_train, y_train)
plot_importance(xbg_model)
pyplot.show()
```

```python
from sklearn.metrics import accuracy_score

predictions = xbg_model.predict(x_test)
acc = accuracy_score(y_test, predictions)
print(str(np.round(acc*100, 2))+'%')
```
